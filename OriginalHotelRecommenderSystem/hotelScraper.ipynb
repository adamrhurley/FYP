```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
from geopy.geocoders import Nominatim
import csv

headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
    }
```

```python
f = open('final-data-Final-batch2.txt', 'r+')
urls = [line for line in f.readlines()]
f.close()
```

```python
    hotel_info_df = pd.DataFrame()
    hotel_info_df['name'] = name
    hotel_info_df['address'] = address
    hotel_info_df['amenities'] = amenities
    hotel_info_df['rooms'] = rooms
    hotel_info_df['rating'] = rating
    hotel_info_df['numReviews'] = numberReviews
    hotel_info_df['lat'] = lat
    hotel_info_df['lon'] = lon
    hotel_info_df.to_csv('hotel_info.csv')
```

```python
numHotels = len(urls)
for i in range(numHotels):
    url = urls[i]
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')

    driver = webdriver.Chrome()
    driver.get(url)

    try:
        name = soup.find(class_="_1mTlpMC3").get_text()
        
    except:
            name = "Unknown"
            print("name not found")
    
    try:
        address = soup.find(class_="jke2_wbp").get_text()
        
    except:
            address = "Unknown"
            print("address not found")
    
    try:
        amenities = driver.find_element_by_xpath("//div[@class='_1nAmDotd'][1]").text
        amenitiesArr = amenities.split("\n")
    
    except:
            amenitiesArr = "Unknown"
            print("amenities not found")
            
    try:
        rooms = driver.find_element_by_xpath("//div[@class='_1nAmDotd'][3]").text
        roomsArr = rooms.split("\n")
                
    except:
            roomsArr = "Unknown"
            print("rooms not found")
            
    try:
        price = driver.find_element_by_xpath("//div[@class='CEf5oHnZ'][1]").text
        
    except:
            price = "Unknown"
            print("price not found")

    try:
        rating = soup.find(class_="_3cjYfwwQ").get_text()

    except:
            rating = "Unknown"
            print("rating not found")
            
    try:        
        numberReviews = soup.find(class_="_3jEYFo-z").get_text()

    except:
            numberReviews = "Unknown"
            print("numberRevies not found")
    try:
            geolocator = Nominatim(user_agent="myGeocoder")
            location = geolocator.geocode(address)
            lat = location.latitude
            lon = location.longitude

    except:
                lat = 'nil'
                lon = 'nil'
                print("Lat & Lon not found")

    
    driver.close()
    with open('hotel_info.csv', 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([name, address, amenitiesArr, roomsArr, price, rating, numberReviews, lat, lon])
    print("Hotel " + str(i) + " Done")
```
